# Seperating the features and labels from dataset

```python
X = df.drop(columns=60,axis=1)
Y = df[60]
```

```python
X
```

```python
Y
```

# Splitting into Training and Testing data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

Now Train the X_train and Y_train data with logistic regression model

```python
model.fit(X_train,Y_train)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score
```

predicting the accuracy of the training data

```python
X_train_pred = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_pred,Y_train)
```

```python
training_data_accuracy
```

```python
print("Accuracy of the Training data is",training_data_accuracy)
```

predicting the accuracy of the testing data

```python
X_test_pred = model.predict(X_test)
```

```python
testing_data_accuracy = accuracy_score(X_test_pred,Y_test)
```

```python
testing_data_accuracy
```

```python
print("Accuracy of the Testing data is",testing_data_accuracy)
```
